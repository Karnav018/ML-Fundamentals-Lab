#### 8. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
arr6 = np.arange(1,10).reshape(3,3)

mean = np.mean(arr6)
std = np.std(arr6)

normalized_array = (arr6 - mean) / std
print("Normalized Array ")
print(normalized_array)
```

#### 10. Create a NumPy array of shape (4, 4) filled with random integers. Subtract a 1D array of shape (4,) from each column of the 2D array using broadcasting.

```python
array = np.random.randint(1, 11, size=(4, 4))
column_array = np.random.randint(1, 11, size=(4,))
print("Original array:")
print(array)
print("1D array:")
print(column_array)

# Subtract the 1D array from each column of the 2D array using broadcasting
result = array - column_array[:, np.newaxis]
print("Resulting array:")
print(result)
```

#### 11. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.

```python
arr7 = np.random.randint(1,11,size=(3,3))
print("Original matrix")
print(arr7)

determinant = np.linalg.det(arr7)
inverse = np.linalg.inv(arr7)
eigenalues = np.linalg.eigvals(arr7)

print("Dterminant : ",determinant)
print("Inverse : ",inverse)
print("Eigen values : ",eigenalues)
```

#### 17. Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.

#### 18. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
# Create a structured array with fields 'x' and 'y'
data_type = [('x', 'i4'), ('y', 'i4')]
data = np.array([(1, 2), (3, 4), (5, 6)], dtype=data_type)
print("Original array:")
print(data)

# Compute the Euclidean distance between each pair of points
distances = np.sqrt((data['x'][:, np.newaxis] - data['x'])**2 + (data['y'][:, np.newaxis] - data['y'])**2)
print("Euclidean distances:")
print(distances)
```

#### 19. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.

```python
import numpy.ma as ma

arr1 = np.random.randint(1,17,size=(4,4))
masked = ma.masked_greater(arr1,10)

print("Original array")
print(arr1)
print("Masked Array")
print(masked)

unmaksed_sum = masked.sum()
print("Sum of unmasked array : ",unmaksed_sum)
```

#### 20. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
import numpy.ma as ma

arr1 = np.random.randint(1,10,size=(3,3))
masked = ma.masked_array(arr1 , mask=np.eye(3,dtype=bool))

print("Original array")
print(arr1)
print("Masked array")
print(masked)

unmasked_mean = masked.mean()
masked = masked.filled(unmasked_mean)
print("Modified masked array")
print(masked)
```
