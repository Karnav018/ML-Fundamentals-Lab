#### 5. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. Introduce some NaN values. Fill the NaN values with the mean of the respective columns.

```python
df = pd.DataFrame(np.random.randint(1,11,size=(5,3)) , columns=['A','B','C'])
print("Original data")
print(df)

# Introducing NaN values
df.iloc[0,2] = np.nan
df.iloc[1,1] = np.nan
df.iloc[4,0] = np.nan
print("Dataframe with NaN values")
print(df)

# Fill the NaN values with the mean of the respective columns
df.fillna(df.mean(), inplace=True)
print("Replacing nan values with mean")
print(df)
```

#### 7. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers. Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.

```python
df = pd.DataFrame({'Category' : np.random.choice(['A','B','C'] , size=10) , 'Value':np.random.randint(1,20,size=10)})
print("Original DataFrame")
print(df)

grouped = df.groupby('Category')['Value'].agg(['sum','mean'])
print("Grouped values")
print(grouped)
```

#### 8. Create a Pandas DataFrame with 3 columns: 'Product', 'Category', and 'Sales'. Fill the DataFrame with random data. Group the DataFrame by 'Category' and compute the total sales for each category.

#### 9. Create two Pandas DataFrames with a common column. Merge the DataFrames using the common column.

```python
df1 = pd.DataFrame({'Key' : ['A','B','C','D'] , 'Value1' : np.random.randint(1,8,size=4)})
df2 = pd.DataFrame({'Key' : ['A','B','C','D'] , 'Value2' : np.random.randint(1,8,size=4)})

print("First Data Frame")
print(df1)

print("Second Data Frame")
print(df2)

merged = pd.merge(df1,df2, on='Key')
print("Merged dataframe")
print(merged)
```

#### 10. Create two Pandas DataFrames with different columns. Concatenate the DataFrames along the rows and along the columns.

```python
df1 = pd.DataFrame({
    'A' : np.random.randint(1,4,size = 4),
    'B' : np.random.randint(1,4,size = 4)
})

df2 = pd.DataFrame({
    'C' : np.random.randint(1,4,size = 4),
    'D' : np.random.randint(1,4,size = 4)
})

print("Data 1 ")
print(df1)
print("Data 2 ")
print(df2)

concat_rows = pd.concat([df1,df2] , axis=0, ignore_index=True)
print("Concatenation by rows")
print(concat_rows)

concat_columns = pd.concat([df1,df2] , axis=1)
print("Concatenation by columns")
print(concat_columns)
```

### Time series analysis

#### 11. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.

```python
date_rng = pd.date_range(start='2025-01-01' , end='2025-03-01', freq='D')
df = pd.DataFrame(date_rng,columns=['date'])
df['data'] = np.random.randint(1,100,size=(len(date_rng)))
df.set_index('date' , inplace=True)
print(df)

monthly_mean = df.resample('M').mean()
print("Monthly mean DataFrame:")
print(monthly_mean)
```

#### 12. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-02-01' and one column filled with random integers. Compute the rolling mean with a window of 7 days.

```python
date_rng = pd.date_range(start='2025-01-01' , end='2025-02-01' , freq='D')
df = pd.DataFrame(date_rng , columns=['date'])
df['data'] = np.random.randint(1,100,size=(len(date_rng)))
df.set_index('date' , inplace=True)
df.head()

# Compute the rolling mean with a window of 7 days
rolling_mean = df.rolling(window=7).mean()
print("Rolling mean dataframe")
print(rolling_mean.head(10))
```

### Multi-Index DataFrame

#### 13. Create a Pandas DataFrame with a MultiIndex (hierarchical index). Perform some basic indexing and slicing operations on the MultiIndex DataFrame.

```python
# Create a Pandas DataFrame with a MultiIndex (hierarchical index)
arrays = [['A', 'A', 'B', 'B'], ['one', 'two', 'one', 'two']]
index = pd.MultiIndex.from_arrays(arrays, names=('Category', 'SubCategory'))
df = pd.DataFrame(np.random.randint(1, 100, size=(4, 3)), index=index, columns=['Value1', 'Value2', 'Value3'])
print("MultiIndex DataFrame:")
print(df)

# Basic indexing and slicing operations
print("Indexing at Category 'A':")
print(df.loc['A'])

print("Slicing at Category 'B' and SubCategory 'two':")
print(df.loc[('B', 'two')])
```
